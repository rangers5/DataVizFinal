```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from plotnine import *
%matplotlib inline
import calendar
from pandas.api.types import CategoricalDtype

#https://www.kaggle.com/datasets/theforcecoder/wind-power-forecasting?resource=download
df = pd.read_csv("Data/cleanData.csv", index_col=0)
df.drop(["month", "day", "year"], axis=1, inplace=True)
```

```python
plotList = []
for col in df.columns[1:]:
    hist = ggplot(df) + geom_histogram(aes(x = col), bins = 20, fill = "grey", color = "black")
    plotList.append(hist)
    ggsave(hist, filename="Plots/Histograms/" + col + ".png")
plotList
```
