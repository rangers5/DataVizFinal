```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from plotnine import *
%matplotlib inline
import calendar
from pandas.api.types import CategoricalDtype

#https://www.kaggle.com/datasets/theforcecoder/wind-power-forecasting?resource=download
df = pd.read_csv("Data/cleanData.csv", index_col=0)
```

### For each column in the cleaned data set other than the date columns, create a box plot to show the distribution and check for outliers

```python
df.drop(["month", "day", "year"], axis=1, inplace=True)
for col in df.columns[1:]:
    plt.subplots()
    plt.boxplot(df[col])
    plt.title(col)    
    fname = "Plots/BoxPlots/" + col + ".png"
    plt.savefig(fname)
    plt.show()
```
